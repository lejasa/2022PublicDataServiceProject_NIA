```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import requests
import pprint
import json
```

```python
url = 'http://apis.data.go.kr/B552474/SenuriService/getJobList?ServiceKey=PYZwqjN5q9aWdK0vWKB3EJF%2FWdLc7ukWq3YzvKgFInDkygXx5vNmBjFL7iK7vMpBxoQP7f5oIOyeId4LQTmRxw%3D%3D&numOfRows=1000&pageNo=1'
result = urlopen(url)
soup = BeautifulSoup(result, 'lxml-xml')
items = soup.find_all('item')
```

```python
# 참고 사이트 : https://operstu1.tistory.com/87
def parse():
    try:
        acptMthd = item.find('acptMthd').get_text()
        deadline = item.find('deadline').get_text()
        emplymShp = item.find('emplymShp').get_text()
        frDd = item.find('frDd').get_text()
        jobId = item.find('jobId').get_text()
        oranNm = item.find('oranNm').get_text()
        organYn = item.find('organYn').get_text()
        recrtTitle = item.find('recrtTitle').get_text()
        stmId = item.find('stmId').get_text()
        stmNm = item.find('stmNm').get_text()
        toDd = item.find('toDd').get_text()
        workPlc = item.find('workPlc').get_text()
        workPlcNm = item.find('workPlcNm').get_text()
        return {
            '접수방법' : acptMthd,
            '접수상태' : deadline,
            '공고형태' : emplymShp,
            '접수시작일' : frDd,
            '공고ID' : jobId,
            '기업명' : oranNm,
            '구분값' : organYn,
            '공고제목' : recrtTitle,
            '시스템ID' : stmId,
            '시스템명' : stmNm,
            '접수종료일' : toDd,
            '근무지' : workPlc,
            '근무지역' : workPlcNm
        }
    except AttributeError as e:
        return {
            '접수방법' : None,
            '접수상태' : None,
            '공고형태' : None,
            '접수시작일' : None,
            '공고ID' : None,
            '기업명' : None,
            '구분값' : None,
            '공고제목' : None,
            '시스템ID' : None,
            '시스템명' : None,
            '접수종료일' : None,
            '근무지' : None,
            '근무지역' : None
        }
```

```python
row = []
for item in items:
    row.append(parse())
```

```python
df = pd.DataFrame(row)
```
