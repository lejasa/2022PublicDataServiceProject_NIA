```python
# 모듈 로딩
import pandas as pd

# 파일 로딩
df = pd.read_csv('./data/지역별디지털배움터.csv')
df
```

```python
# 행 기준 중복 제거
df = df.drop_duplicates()
```

```python
# 인덱스 초기화
df = df.reset_index(drop=True)
```

```python
# '장소/위치' 컬럼 값 이상하여 데이터 전처리
result = []

for idx in range(df.shape[0]):
    value = df.loc[idx]['장소/위치'].split()
    for i in range(len(value)):
        if ('길' in value[i])|('로' in value[i]):
            result.append(' '.join(value[:i+2]))
            break
```

```python
# '장소/위치' 데이터 새로운 값 넣기
df['장소/위치'] = pd.Series(result)
```

```python
# 데이터 프레임 df <- csv 파일로 저장
df.to_csv('지역별디지털배움터.csv', index=False)
```
