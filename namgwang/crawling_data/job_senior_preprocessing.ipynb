```python
# 모듈 로딩
import os
import pandas as pd

# 경로 파일명
DIR = os.listdir('./data/data')

df_list=[]  # 하나의 데이터프레임을 합치기 위해 담을 리스트

# 파일 로딩 후 리스트에 담기
for file in DIR:
    df_list.append(pd.read_csv('./data/data'+file))
```

```python
# 하나의 데이터프레임으로 병합
df = pd.concat(df_list)
df
```

```python
# 'c' 컬럼 값에서 마감 제거
df1 = df[df['c'] != '마감']

# 'c' 컬럼 값 확인
df1['c'].value_counts()
```

```python
# 필요 없는 컬럼 지우기
df2 = df1.drop(['Unnamed: 0', '    '], axis=1)
```

```python
# 컬럼명 수정
df2.columns=['지역 및 시군구','기업명','접수상태','주소','연락처','구인사항','접수마감일자']
```

```python
# 행 기준 중복 제거
DF = df2.drop_duplicates()
```

```python
# 인덱스 초기화
DF1 = DF.reset_index(drop=True)
DF1
```

```python
# 컬럼명 확인
DF1.columns
```

```python
sd=[]  # 시도 담을 리스트
sgg=[]  # 시군구 담을 리스트

for idx in DF1.index:
    sd.append(DF1.iloc[idx]['지역 및 시군구'].split()[0])
    sgg.append(DF1.iloc[idx]['지역 및 시군구'].split()[1])
```

```python
# '시도' 컬럼 삽입
DF1.insert(0, '시도', sd)
```

```python
# '시군구' 컬럼 삽입
DF1.insert(1, '시군구', sgg)
```

```python
# 데이터프레임 DF1 -> csv 파일로 저장
DF1.to_csv('일자리끝.csv', index=False)
```
