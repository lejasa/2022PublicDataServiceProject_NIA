```python
# 모듈 로딩
import pandas as pd

# 파일 로딩
df = pd.read_excel('./data/노인맞춤돌봄서비스_수행기관현황.xlsx', header=3)
df
```

```python
# 컬럼명 확인
df.columns
```

```python
# 필요 없는 컬럼 삭제
df = df.drop(['Unnamed: 0', '특화서비스'], axis = 1)
df
```

```python
# '시군구' 컬럼 값 확인
df.시군구.unique()
```

```python
# '시군구' 컬럼 값 '*광역수행기관' 이면 해당 인덱스 '시도' 컬럼 값으로 변경
for idx in range(df.shape[0]):
    if df.iloc[idx]['시군구'] == '*광역수행기관':
        df.iloc[idx]['시군구'] = df.iloc[idx]['시도']
```

```python
# '시군구' 컬럼 값 확인
df.시군구.unique()
```

```python
# 데이터프레임 df <- csv 파일로 저장
df.to_csv('노인맞춤돌봄서비스_수행기관현황.csv', index=False)
```
